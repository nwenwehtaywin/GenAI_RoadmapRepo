```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis") # default
classifier("i hate this movie. The plot is not my taste.")
```

```python
inputs=["I don't think I like this haircut.",
        "I like ordering pizza, but not sure.",
        "I am not sure I like this movie"]
classifier(inputs)
```
